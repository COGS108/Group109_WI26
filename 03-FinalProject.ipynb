#### Cleaning 
To address missingness, we narrowed the dataset to the three columns relevant to our research question, Apprehension Date, Citizenship Country, and Anonymized Identifier, excluding all columns related to case outcomes, departure, and geographic processing since our analysis only measures arrest counts by country and year. Rows missing an Anonymized Identifier were dropped because without one there is no reliable way to distinguish individual records across the two datasets, risking double-counting or incorrect matches when merging. Rows missing an Apprehension Date were also dropped since date is a core component of our analysis, without it we cannot track how arrests change over time. Finally, Apprehension Date was converted to date-only format, dropping the time component, and all retained columns were renamed to match the conventions used in the first dataset to ensure consistency before merging.

```python
ice_removals_2011_2023_clean = ice_removals_2011_2023.dropna(subset=['Apprehension Date','Anonymized Identifer']).copy()
ice_removals_2011_2023_clean['Apprehension Date'] = pd.to_datetime(ice_removals_2011_2023_clean['Apprehension Date']).dt.strftime('%Y-%m-%d')
ice_removals_2011_2023_clean = ice_removals_2011_2023_clean.rename(columns={
    'Apprehension Date': 'apprehension_date',
    'Citizenship Country': 'citizenship_country',
    'Anonymized Identifer': 'unique_identifier'
})
ice_removals_2011_2023_clean = ice_removals_2011_2023_clean[['apprehension_date','citizenship_country','unique_identifier']]
ice_removals_2011_2023_clean.head()
```

### Joining Datasets
As noted earlier, the ICE 2023–2025 Arrests Dataset only covers September 2023 through October 2025, meaning arrests before that window are not captured. To ensure a more complete picture, we supplemented it with records from the ICE 2011–2023 Removals Dataset, filtering to only include entries after January 1, 2023 to avoid overlap. Although the Removals dataset primarily tracks deportation outcomes rather than arrests, it contains an apprehension date field, the same key variable used in the Arrests dataset. This means the Removals dataset effectively captures a subset of arrests, specifically those that ultimately led to a removal. By filtering on apprehension date, we are able to use these records to fill in arrest activity from early 2023 that the Arrests dataset does not cover, while acknowledging that this subset excludes arrests that did not result in removal. After cleaning, both datasets were aligned to share the same column names and structure, making pd.concat a straightforward way to combine them into a single unified dataset. Following the merge, duplicate rows were dropped to ensure each arrest record appears only once in the final dataset.

```python
ice_removals_2023 = ice_removals_2011_2023_clean[ice_removals_2011_2023_clean['apprehension_date']>='2023-02-01']
ice_2023_2025 =pd.concat([ice_arrests_2023_2025_clean, ice_removals_2023], ignore_index=True)
ice_2023_2025 = ice_2023_2025.drop_duplicates()
ice_2023_2025.sort_values(by='apprehension_date').head()
```

#### Section 1: Analyzing Years
Before examining patterns across countries, it is worth first looking at the total number of arrests recorded each `year`, as this provides important context for the proportions calculated later. To accomplish this, we created a new year column derived from the `apprehension_date` column, which tells us the year each arrest took place. With that column in place, we then grouped by `year` and applied the aggregate function `COUNT` to get the total number of arrests for each year.

```python
ice_2023_2025['year'] = pd.to_datetime(ice_2023_2025['apprehension_date']).dt.year
ice_2023_2025 = ice_2023_2025[ice_2023_2025['year'].isin([2023, 2024, 2025])]
arrests_per_year = ice_2023_2025.groupby('year')['unique_identifier'].count()
arrests_per_year.plot(kind='bar', figsize=(8, 5))
plt.title('Total ICE Arrests per Year')
plt.xlabel('Year')
plt.ylabel('Number of Arrests')
```

Despite this shorter timeframe, 2025 already far exceeds the totals of both previous years, suggesting a significant increase in ICE arrest activity.

This is worth keeping in mind when interpreting our results because 2025 has a much larger total number of arrests, a country could have the same number of arrests as prior years but still appear to have a smaller share, simply because the overall pool is bigger. This means any drop in a country's share in 2025 may not necessarily reflect fewer arrests, but rather that arrests grew faster overall.


#### Section 2: Analyzing Proportions
Moving on to analyzing proportions, we can answer the research question: How has the distribution of ICE arrests by country of citizenship changed across 2023, 2024, and 2025? To do this, we first need to calculate the total number of arrests per country per year. We apply the aggregate function `COUNT` on the `unique_identifier` column, grouped by `year` and `citizenship_country`, which gives us the raw arrest counts. From there, we can calculate each country's proportion of total arrests within a given year by dividing the raw arrest count by the number of arrests that took place that year. Allowing for a meaningful year-over-year comparison of the distribution.

One potentially confusing pattern in our results is the slight decline in Mexico's share of arrests from 2024 (~41%) to 2025 (~38%), despite 2025 having by far the highest total arrest count of the three years. At first glance, this might suggest reduced enforcement against Mexican nationals, which would contradict the overall trend of escalating ICE activity. As the total number of arrests grew dramatically in 2025, arrests of nationals from other countries, particularly Venezuela, grew at a faster rate, diluting Mexico's proportional share even while its absolute numbers remained high. Similarly, Nicaragua's appearance in the 2024 top 5 followed by its disappearance in 2025 could seem inconsistent, but likely reflects a temporary migration surge rather than a systematic enforcement shift. These fluctuations are best understood not as contradictions but as expected noise within an otherwise stable pattern of Latin American overrepresentation.

5. Several limitations affect how confidently we can interpret these findings.
First, the dataset is not a fully unbiased sample. The 2023–2025 Arrests dataset begins only in September 2023, so early 2023 observations rely on the Removals dataset, which includes only individuals who were ultimately deported. This likely undercounts arrests in early 2023 and may make that year appear artificially lower in comparison. Second, the 2025 data only runs through October, making direct yearly comparisons difficult without adjusting for the shorter observation period. Third, our analysis relies on cross-sectional snapshots rather than longitudinal data, which limits our ability to track how enforcement patterns evolve in response to policy changes or migration events. Fourth, enforcement patterns appear to shift over time, as seen in the surge of arrests in 2025 and the changing presence of certain countries in the top five. Finally, the analysis does not normalize arrest shares by the estimated undocumented population from each country, making it difficult to determine whether overrepresentation reflects enforcement patterns or population size.
