# Research Question

How has the distribution of ICE arrests by country of citizenship changed across 2023, 2024, and 2025, and are certain countries consistently over- or underrepresented relative to their share of total arrests?

## Background and Prior Work

ICE is the U.S. Immigration and Customs Enforcement, or an agency within the Department of Homeland Security responsible for deportation and investigations regarding immigration law violations. After 9/11, the Homeland Security Act of 2002 was established. This act was established to reduce threats by terrorist networks and enforce immigration laws. In the following year, 2003, ICE was created under the Department of Homeland Security. Although ICE was created to enforce immigration laws, it has become very controversial in recent years, as some claim ICE has allegedly over-discriminated certain racial groups.

Previously, work regarding our research question has been done and public data sets exist that compile ICE arrest and deportation records.<a href="#ref1">1</a> This dataset shows information, including race, on many people ICE has encountered, arrested, detained, transported, deported, and jailed. However, this dataset lacks information about the total amount of immigrants from each country, and just gives raw data about the amount of immigrants that were arrested, etc. from a given country. There has also been research done regarding ICE arrest statistics in major US cities and their different types of arrests. <a href="#ref2">2</a> This source has data on the arrest date, arrest city, country of citizenship, and criminality. However, a drawback to this source is the lack of accessibility of more specific data. There is only broad statistics having to do with ICE arrests in major cities. Furthermore, there is information about individual encounters with ICE enforcement in the final source. <a href="#ref3">3</a> This source gives much more individual detail than the other sources, but still is lacking in terms of real-time updates.

Through analyzing data from these three sources, we can begin to answer how the distribution of ICE arrests by country of citizenship has shifted across 2023, 2024, and 2025, and whether certain countries have become more or less overrepresented over time. Specifically, our group is measuring arrests done by ICE per country of origin of immigrants in the United States. This will be represented by the number of immigrants from a given country that have been arrested divided by the total number of ICE cases in the from the respective year 2023, 2024 and 2025.

Key Metrics:

Apprehension Date (date/timestamp format): This variable records the date on which the enforcement action was recorded. The unit of measurement is calendar time at the day level. For analytical purposes, we extract the year from this variable to aggregate removals by calendar year and enable longitudinal comparisons across 2011–2023.

Country of Citizenship (categorical variable): This variable identifies the nationality of the individual removed. It is central to our research question, as we calculate the proportion of total removals attributable to each country in a given year.

Arrest Count (derived metric): Although the dataset does not directly report overall totals, each row represents a single removal case. Therefore, counting the number of rows gives the total number of removals. We can group these rows (for example, by year or country) to calculate counts and then convert those counts into proportions, such as each country’s share of total annual removals. Furthermore, although the dataset is structured around removals, the inclusion of the Apprehension Date column allows us to focus on the timing of apprehensions and treat each row as representing a single arrest case. In this way, we can similarly compute totals and proportions for arrests by year or country.

Primary Outcome of Interest: is the proportional composition of arrests by country within each year.
Because overall enforcement volume fluctuates across years, we focus on proportions rather than raw counts. For example, if 30% of removals in a given year involve individuals from a specific country, this indicates that nearly one-third of arrests that year were associated with that nationality. This proportional approach allows us to compare enforcement composition over time even when total arrests increase or decrease.

2B: Policy-driven variation: ICE enforcement priorities shifted across administrations between 2011 and 2023. As a result, observed changes in arrest proportions may reflect policy changes rather than underlying migration flows.
Administrative data bias: This dataset captures enforcement outcomes, not the total undocumented population. It measures government enforcement activity rather than population-level prevalence.
Missing administrative identifiers: Some internal identifiers contain missing values. However, because our analysis depends primarily on apprehension date and country of citizenship, this missingness does not materially affect our proportional calculations.
Scope limitation: The dataset includes only individuals who were formally removed. It excludes those apprehended but released, voluntarily departed, or never encountered by enforcement authorities. Therefore, the data reflect enforcement outcomes rather than the broader undocumented population. ### Joining Datasets
As noted earlier, the ICE 2023–2025 Arrests Dataset only covers September 2023 through October 2025, meaning arrests before that window are not captured. To ensure a more complete picture, we supplemented it with records from the ICE 2011–2023 Removals Dataset, filtering to only include entries after January 1, 2023 to avoid overlap. Although the Removals dataset primarily tracks deportation outcomes rather than arrests, it contains an apprehension date field, the same key variable used in the Arrests dataset. This means the Removals dataset effectively captures a subset of arrests, specifically those that ultimately led to a removal. By filtering on apprehension date, we are able to use these records to fill in arrest activity from early 2023 that the Arrests dataset does not cover, while acknowledging that this subset excludes arrests that did not result in removal. #### Section 2: Analyzing Proportions
Moving on to analyzing proportions, we can answer the research question: How has the distribution of ICE arrests by country of citizenship changed across 2023, 2024, and 2025? To do this, we first need to calculate the total number of arrests per country per year. We apply the aggregate function `COUNT` on the `unique_identifier` column, grouped by `year` and `citizenship_country`, which gives us the raw arrest counts. From there, we can calculate each country's proportion of total arrests within a given year by dividing the raw arrest count by the number of arrests that took place that year. Allowing for a meaningful year-over-year comparison of the distribution.

```python
arrests_by_country_year = ice_2023_2025.groupby(['year', 'citizenship_country'])['unique_identifier'].count().reset_index()
arrests_by_country_year.columns = ['year', 'citizenship_country', 'arrest_count']
yearly_totals = arrests_by_country_year.groupby('year')['arrest_count'].transform('sum')
arrests_by_country_year['share'] = arrests_by_country_year['arrest_count'] / yearly_totals * 100
top5 = (arrests_by_country_year.groupby('year')
          .apply(lambda x: x.nlargest(5, 'share'))
          .reset_index(drop=True))

years = sorted(top5['year'].unique())
fig, axes = plt.subplots(1, len(years), figsize=(16, 6), sharey=False)

for ax, year in zip(axes, years):
    data = top5[top5['year'] == year].sort_values('share', ascending=False)
    ax.bar(data['citizenship_country'], data['share'], color='steelblue', edgecolor='white')
    ax.set_title(f'{year}', fontsize=14, fontweight='bold')
    ax.set_xlabel('Country')
    ax.set_ylabel('Share (%)')
    ax.tick_params(axis='x', rotation=45)

plt.suptitle('Top 5 Countries by Share of ICE Arrests (2023–2025)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig("results/share_ice_arrests.png", dpi=300, bbox_inches="tight")
plt.show()
```

The graph displays three side-by-side bar charts, each showing the top 5 countries by share of ICE arrests for a given year. Honduras and Guatemala remain fixtures in the top 5 across all three years, though their individual shares fluctuated slightly. Colombia dropped out of the top 5 after 2023, suggesting a decline in arrests of Colombian nationals relative to other groups.Nicaragua appeared in 2024 but exited the top 5 by 2025, indicating a temporary spike.Venezuela entered the top 5 in 2025, which aligns with broader migration trends driven by the ongoing political and economic crisis in Venezuela. Overall, across all three years, the top 5 countries by share of ICE arrests were exclusively Latin American nations, specifically from Central America, South America, and Mexico. This consistent pattern suggests that Latin American citizens make up a disproportionately large share of ICE arrests, and despite some shuffling in which specific countries appear in the top 5 each year, the regional dominance of Latin America remained unchanged throughout 2023, 2024, and 2025

#### Section 3: Overrepresentation through boostrap

Although the previous section has already alluded to Mexico's consistent overrepresentation in ICE arrests, we further illustrate the magnitude of this pattern through bootstrap resampling. To do this, we repeatedly drew 1,000 random samples of 500 arrests from each year's data, calculated each country's share of arrests in every sample, and used the resulting distribution to construct 95% confidence intervals around each country's mean share. 

```python
def bootstrap_shares(df, year, n_bootstrap=1000, sample_size=500):
    year_data = df[df['year'] == year]
    shares = []
    
    for _ in range(n_bootstrap):
        sample = year_data.sample(n=sample_size, replace=True)
        share = sample['citizenship_country'].value_counts(normalize=True) * 100
        shares.append(share)
    
    shares_df = pd.DataFrame(shares).fillna(0)
    
    results = pd.DataFrame({
        'mean_share': shares_df.mean(),
        'ci_lower': shares_df.quantile(0.025), 
        'ci_upper': shares_df.quantile(0.975)
    }).sort_values('mean_share', ascending=False)
    
    return results

results_2023 = bootstrap_shares(ice_2023_2025, 2023)
results_2024 = bootstrap_shares(ice_2023_2025, 2024)
results_2025 = bootstrap_shares(ice_2023_2025, 2025)

fig, axes = plt.subplots(1, 3, figsize=(20, 6), sharey=True)
results = [results_2023, results_2024, results_2025]
years = [2023, 2024, 2025]

for ax, result, year in zip(axes, results, years):
    top10 = result.head(10)
    ax.barh(top10.index, top10['mean_share'], color='#1a80bb', alpha=0.7)
    ax.errorbar(
        top10['mean_share'], top10.index,
        xerr=[top10['mean_share'] - top10['ci_lower'],
              top10['ci_upper'] - top10['mean_share']],
        fmt='none', color='black', capsize=4 
    )
    ax.set_title(f'{year}', fontsize=13, fontweight='bold')
    ax.set_xlabel('Mean Share (%) with 95% CI')

fig.suptitle('Bootstrap Estimated Share of ICE Arrests by Country (2023–2025)',
             fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig("results/bootstraped.png", dpi=300, bbox_inches="tight")
plt.show()
```

Each horizontal bar represents a country's mean share of ICE arrests estimated across 1,000 bootstrap samples, with the black error bars showing the 95% confidence interval. The top 10 countries are shown per year, ordered from lowest to highest share. The top 10 countries are shown per year, ordered from lowest to highest share. 

Mexico has by far the longest bar in all three years, with a mean share of approximately ~32% in 2023, ~41% in 2024, and ~38% in 2025.Guatemala and Honduras consistently appear as the second and third highest countries across all three years with shares around 10–16%
The remaining countries, Colombia, El Salvador, Ecuador, Peru, Nicaragua, Venezuela, Dominican Republic, and Cuba,  all have notably shorter bars, clustering between 1–6%

Mexico's overrepresentation is unambiguous, its bar dwarfs every other country in all three years, and its narrow confidence intervals confirm this is not a statistical anomaly but a consistent and reliable pattern. Guatemala and Honduras are also clearly overrepresented, with mean shares significantly higher than the remaining countries and relatively tight confidence intervals confirming their elevated shares are stable across all three years. The fact that all top 10 countries are Latin American across every year further reinforces the regional concentration of ICE arrests identified throughout this analysis Overall, the bootstrap results provide strong statistical evidence that Mexico, Guatemala, and Honduras are not just frequently appearing in the data, they are systematically and significantly overrepresented in ICE arrests relative to all other nations
