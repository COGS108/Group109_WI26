## Abstract

<!-- Please write one to four paragraphs that describe a very brief overview of why you did this, how you did, and the major findings and conclusions. -->

This study examines patterns in U.S. Immigration and Customs Enforcement (ICE) arrests by country of citizenship from 2023 to 2025. Our group chose this topic because it is both interesting and highly relevant to people living in the United States today, particularly given the ongoing public attention surrounding immigration enforcement. The primary research question is: How has the distribution of ICE arrests by countries of citizenship changed over time, and do these distributions exhibit consistent patterns over time?

The analysis uses publicly available enforcement datasets from ICE and Kaggle, which provide aggregate statistics on arrests and removals without including personally identifiable information. Data were cleaned, merged, and organized to enable consistent comparisons across years. We computed yearly totals and country-level arrest shares and used visualizations and summary statistics to compare patterns across nationalities. Additionally, bootstrap resampling was applied to estimate mean arrest shares and construct 95% confidence intervals, allowing us to assess the stability of observed trends.

Our findings show that immigrants from Latin American countries consistently account for the largest share of ICE arrests across all three years, with Mexico, Guatemala, and Honduras most prominently represented. While total arrest counts increased over time, the distribution of arrests changed to a moderate extent. Although Latin American countries remain overrepresented overall, specific countries experienced noticeable shifts in their share of total arrests, reflecting a broader pattern in which some nationalities are becoming more prominent while others decline or fluctuate. Overall, the results indicate that ICE enforcement activity is not evenly distributed across countries of citizenship but instead reflects a persistent, yet gradually evolving, geographic concentration.

## Background and Prior Work

ICE is the U.S. Immigration and Customs Enforcement, or an agency within the Department of Homeland Security responsible for deportation and investigations regarding immigration law violations. After 9/11, the Homeland Security Act of 2002 was established. This act was established to reduce threats by terrorist networks and enforce immigration laws. In the following year, 2003, ICE was created under the Department of Homeland Security. Although ICE was created to enforce immigration laws, it has become very controversial in recent years, as some claim ICE has allegedly over-discriminated certain racial groups.

Previously, work regarding our research question has been done and public data sets exist that compile ICE arrest and deportation records.<a href="#ref1">1</a> This dataset shows information, including race, on many people ICE has encountered, arrested, detained, transported, deported, and jailed. However, this dataset lacks information about the total amount of immigrants from each country, and just gives raw data about the amount of immigrants that were arrested, etc. from a given country. There has also been research done regarding ICE arrest statistics in major US cities and their different types of arrests. <a href="#ref2">2</a> This source has data on the arrest date, arrest city, country of citizenship, and criminality. However, a drawback to this source is the lack of accessibility of more specific data. There is only broad statistics having to do with ICE arrests in major cities. Furthermore, there is information about individual encounters with ICE enforcement in the final source. <a href="#ref3">3</a> This source gives much more individual detail than the other sources, but still is lacking in terms of real-time updates.

Through analyzing data from these three sources, we can begin to answer how the distribution of ICE arrests by country of citizenship has shifted across 2023, 2024, and 2025, and whether certain countries have become more or less overrepresented over time.

#### Cleaning 
To address missingness, we narrowed the dataset to the three columns relevant to our research question, Apprehension Date, Citizenship Country, and Anonymized Identifier, excluding all columns related to case outcomes, departure, and geographic processing since our analysis only measures arrest counts by country and year. Rows missing an Anonymized Identifier were dropped because without one there is no reliable way to distinguish individual records across the two datasets, risking double-counting or incorrect matches when merging. Rows missing an Apprehension Date were also dropped since date is a core component of our analysis, without it we cannot track how arrests change over time. Finally, Apprehension Date was converted to date-only format, dropping the time component, and all retained columns were renamed to match the conventions used in the first dataset to ensure consistency before merging.

```python
# Drop rows missing apprehension date or anonymized ID, and make an explicit copy to avoid modifying the original
ice_removals_2011_2023_clean = ice_removals_2011_2023.dropna(subset=['Apprehension Date','Anonymized Identifer']).copy()
# Standardize the date format to YYYY-MM-DD
ice_removals_2011_2023_clean['Apprehension Date'] = pd.to_datetime(ice_removals_2011_2023_clean['Apprehension Date']).dt.strftime('%Y-%m-%d')
# Rename columns to match the naming convention used in the arrests dataset
ice_removals_2011_2023_clean = ice_removals_2011_2023_clean.rename(columns={
    'Apprehension Date': 'apprehension_date',
    'Citizenship Country': 'citizenship_country',
    'Anonymized Identifer': 'unique_identifier'
})
# Keep only the three relevant columns
ice_removals_2011_2023_clean = ice_removals_2011_2023_clean[['apprehension_date','citizenship_country','unique_identifier']]
ice_removals_2011_2023_clean.head()
```

### Joining Datasets
As noted earlier, the ICE 2023–2025 Arrests Dataset only covers September 2023 through October 2025, meaning arrests before that window are not captured. To ensure a more complete picture, we supplemented it with records from the ICE 2011–2023 Removals Dataset, filtering to only include entries after January 1, 2023 to avoid overlap. Although the Removals dataset primarily tracks deportation outcomes rather than arrests, it contains an apprehension date field, the same key variable used in the Arrests dataset. This means the Removals dataset effectively captures a subset of arrests, specifically those that ultimately led to a removal. By filtering on apprehension date, we are able to use these records to fill in arrest activity from early 2023 that the Arrests dataset does not cover, while acknowledging that this subset excludes arrests that did not result in removal. After cleaning, both datasets were aligned to share the same column names and structure, making pd.concat a straightforward way to combine them into a single unified dataset. Following the merge, duplicate rows were dropped to ensure each arrest record appears only once in the final dataset.

```python
# Filter the removals dataset to only include records from February 2023 onward
ice_removals_2023 = ice_removals_2011_2023_clean[ice_removals_2011_2023_clean['apprehension_date']>='2023-01-01']
# Combine the 2023-onward removals with the arrests dataset into a single dataframe
ice_2023_2025 =pd.concat([ice_arrests_2023_2025_clean, ice_removals_2023], ignore_index=True)
# Remove any duplicate records that may have appeared in both datasets
ice_2023_2025 = ice_2023_2025.drop_duplicates()
# Preview the earliest records by sorting chronologically
ice_2023_2025.sort_values(by='apprehension_date')
```

#### Section 1: Analyzing Years
Before examining patterns across countries, it is worth first looking at the total number of arrests recorded each `year`, as this provides important context for the proportions calculated later. To accomplish this, we created a new year column derived from the `apprehension_date` column, which tells us the year each arrest took place. With that column in place, we then grouped by `year` and applied the aggregate function `COUNT` to get the total number of arrests for each year.

```python
# Extract the year from the apprehension date into a new column
ice_2023_2025['year'] = pd.to_datetime(ice_2023_2025['apprehension_date']).dt.year
# Filter to only include records from 2023, 2024, and 2025 (removes any outlier dates)
ice_2023_2025 = ice_2023_2025[ice_2023_2025['year'].isin([2023, 2024, 2025])]
# Count the number of arrests/removals per year using unique_identifier as the row counter
arrests_per_year = ice_2023_2025.groupby('year')['unique_identifier'].count()
# Plot the yearly counts as a bar chart
arrests_per_year.plot(kind='bar', figsize=(8, 5))
plt.title('Total ICE Arrests per Year')
plt.xlabel('Year')
plt.ylabel('Number of Arrests')
```

<center><b>Figure 1</b></center>


Looking at Figure 1,the total number of arrests per year, 2023 and 2024 recorded a similar amounts of arrests. Notably, 2025 has the highest count, this is particularly striking given that the dataset only covers through October 2025, meaning 2025 represents less than a full year of data. Despite this shorter timeframe, 2025 already far exceeds the totals of both previous years, suggesting a significant increase in ICE arrest activity.

This is worth keeping in mind when interpreting our results because 2025 has a much larger total number of arrests, a country could have the same number of arrests as prior years but still appear to have a smaller share, simply because the overall pool is bigger. This means any drop in a country's share in 2025 may not necessarily reflect fewer arrests, but rather that arrests grew faster overall.

#### Section 3: Overrepresentation through boostrap
Although the previous section has already alluded to Mexico's consistent overrepresentation in ICE arrests, we further illustrate the magnitude of this pattern through bootstrap resampling. To do this, we repeatedly drew 1,000 random samples of 500 arrests from each year's data, calculated each country's share of arrests in every sample, and used the resulting distribution to construct 95% confidence intervals around each country's mean share. 

##### To see the code responsible for Figure 4 analyses see 02-EDACheckpoint.ipynb

```python
Image("results/bootstraped.png")
```

<center><b>Figure 4</b></center>

Each horizontal bar in Figure 4 represents a country's mean share of ICE arrests estimated across 1,000 bootstrap samples, with the black error bars showing the 95% confidence interval. The key value of this approach is that it moves beyond simply observing that certain countries appear frequently in the data by repeatedly resampling the dataset, we can quantify how reliably and consistently each country's share holds up, and determine whether the overrepresentation we observe is a genuine pattern or a product of random variation.

The results are pretty obvious. Mexico's mean share of approximately 32% in 2023, 41% in 2024, and 38% in 2025 is accompanied by notably narrow confidence intervals across all three years. This is statistically significant as narrow intervals mean that no matter which 500 arrests were drawn in any given sample, Mexico's share remained consistently high. If Mexico's dominance were a coincidence or an artifact of how the data was collected, we would expect wide, unstable intervals. Instead, the tightness of the intervals confirms that Mexico's overrepresentation is a structural feature of ICE arrests, not a statistical anomaly.

Guatemala and Honduras tell a similar story. Their consistently elevated mean shares, paired with tight confidence intervals across all three years, confirm that their overrepresentation is equally stable and reliable. The remaining countries: Colombia, El Salvador, Ecuador, Peru, Nicaragua, Venezuela, Dominican Republic, and Cuba cluster between 1% and 6% with wider intervals, indicating both lower representation and greater variability in their shares across samples.

Perhaps the most striking finding is that all top 10 countries across every year are Latin American. The bootstrap results do not just confirm which countries are overrepresented, they confirm that this regional concentration is statistically robust and consistent, providing strong evidence that the distribution of ICE arrests is systematically skewed toward Latin American nations, and has remained so throughout 2023, 2024, and 2025.
